```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# ADRO

```python
lstm_adro = "https://raw.githubusercontent.com/MIlhamEr/Skripsi/refs/heads/main/LSTM_adro_predictions.csv"
lstm_adro = pd.read_csv(lstm_adro)
lstm_adro.tail()
```

```python
gru_adro = "https://raw.githubusercontent.com/MIlhamEr/Skripsi/refs/heads/main/GRU_adro_predictions.csv"
gru_adro = pd.read_csv(gru_adro)
gru_adro.tail()
```

```python
gru_adro = gru_adro.drop(columns=["Actual ADRO"])
gru_adro.rename(columns={'Predicted ADRO': 'Predicted ADRO GRU'}, inplace=True)
```

```python
merged_adro = pd.concat([lstm_adro, gru_adro], axis=1)
merged_adro.tail()
```

# DSSA

```python
lstm_dssa = "https://raw.githubusercontent.com/MIlhamEr/Skripsi/refs/heads/main/LSTM_dssa_predictions.csv"
lstm_dssa = pd.read_csv(lstm_dssa)
lstm_dssa.tail()
```

```python
gru_dssa = "https://raw.githubusercontent.com/MIlhamEr/Skripsi/refs/heads/main/GRU_dssa_predictions.csv"
gru_dssa = pd.read_csv(gru_dssa)
gru_dssa.tail()
```

```python
gru_dssa = gru_dssa.drop(columns=["Actual DSSA"])
gru_dssa.rename(columns={'Predicted DSSA': 'Predicted DSSA GRU'}, inplace=True)
```

```python
merged_dssa = pd.concat([lstm_dssa, gru_dssa], axis=1)
merged_dssa.tail()
```

# Plot the predicted vs actual values

```python
merged_adro.tail()
```

```python
merged_dssa.tail()
```

```python
plt.figure(figsize=(16, 4), facecolor="#f8f4f4")
plt.plot(merged_adro['Actual ADRO'], label='Actual')
plt.plot(merged_adro['Predicted ADRO'], label='LSTM Predicted')
plt.plot(merged_adro['Predicted ADRO GRU'], label='GRU Predicted')
plt.title("Prediction for ADRO.JK")
plt.legend()
plt.grid(True, linestyle="--", linewidth=0.5)
plt.gca().set_facecolor("#f8f4f4")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 4), facecolor="#f8f4f4")
plt.plot(merged_dssa['Actual DSSA'], label='Actual')
plt.plot(merged_dssa['Predicted DSSA'], label='LSTM Predicted')
plt.plot(merged_dssa['Predicted DSSA GRU'], label='GRU Predicted')
plt.title("Prediction for DSSA.JK")
plt.legend()
plt.grid(True, linestyle="--", linewidth=0.5)
plt.gca().set_facecolor("#f8f4f4")
plt.tight_layout()
plt.show()
```
